## Import

```python
%matplotlib inline
import matplotlib.pyplot as plt
import os, sys
from PIL import Image

from utils.helpers import *

from sklearn import linear_model
```

## Load the data

```python
root_training_dir = "dataset/training/"
image_dir = root_training_dir + "images/"
gt_dir = root_training_dir + "groundtruth/"
root_test_dir = "dataset/test_set_images/"


files_train = os.listdir(image_dir)
folder_test = os.listdir(root_test_dir)

n_train = len(files_train)
n_test = len(folder_test)


imgs = [load_image(image_dir + files_train[i]) for i in range(n_train)]
train_arrays = [np.array(img) / 255 for img in imgs]

gt_imgs = [load_image(gt_dir + files_train[i]) for i in range(n_train)]
train_gt_arrays = [np.array(img) / 255 for img in gt_imgs]


imgs_test = [ load_image(os.path.join(root_test_dir, folder_test[i], f"{folder_test[i]}.png")) for i in range(n_test)]
imgs_test_arrays = [np.array(img) / 255 for img in imgs_test]
```

```python
# Extract patches from input images
patch_size = 16  # each patch is 16*16 pixels

img_patches = [img_crop(imgs[i], patch_size, patch_size) for i in range(n_train)]
gt_patches = [img_crop(gt_imgs[i], patch_size, patch_size) for i in range(n_train)]

# Convert to numpy arrays
img_patches = np.array(img_patches)
gt_patches = np.array(gt_patches)

print(f"Shape of unflattened image patches : {img_patches.shape}\n"
      f"Shape of unflattened ground truth patches : {gt_patches.shape} \n")

# Linearize list of patches
img_patches = np.asarray(
    [
        img_patches[i][j]
        for i in range(len(img_patches))
        for j in range(len(img_patches[i]))
    ]
)
gt_patches = np.asarray(
    [
        gt_patches[i][j]
        for i in range(len(gt_patches))
        for j in range(len(gt_patches[i]))
    ]
)
print(f"Shape of flattened image patches : {img_patches.shape}\n"
      f"Shape of flattened ground truth patches : {gt_patches.shape}")
```

## Feature extraction
In order to build features for the logistic regression, we can work with different features extration methods, indeed we can extract:
- 2 features: The mean and std of grey color
- 6 features: The mean and std for the three RGB colors: 2*3=6

- What if we try 8 features by summing the two methods?

```python
X2d = np.asarray([extract_features_2d(img_patches[i]) for i in range(len(img_patches))])
X6d = np.asarray([extract_features(img_patches[i]) for i in range(len(img_patches))])
X8d = np.asarray([np.concatenate((X2d[i], X6d[i])) for i in range(len(X2d))])

y = np.asarray([value_to_class(np.mean(gt_patches[i]),foreground_threshold = 0.25) for i in range(len(gt_patches))])

print("length of X2d, X6d and X8d = {},{} and {}".format(len(X2d),len(X6d),len(X8d)))
print("Feature dimension for X2d= " + str(X2d.shape[1]))
print("Feature dimension for X6d= " + str(X6d.shape[1]))
print("Feature dimension for X8d= " + str(X8d.shape[1]))
print("Number of classes = " + str(np.max(y)))
```

For visualization, let's plot the distribution of road vs non-road patches (the first 1000 patchs for clarity) for the simple case of 2D features (grey mean and variance)

```python
# Plot 2d features using groundtruth to color the datapoints
plt.scatter(X2d[:1000, 0], X2d[:1000, 1], c=y[:1000], edgecolors="k", cmap=plt.cm.Paired)

plt.title("Distribution of road vs non-road patches")
plt.xlabel("Average grey value of the pixels in the patch")
plt.ylabel("Variance of grey value of the pixels in the patch")
plt.show()
```

```python
# split the data into train and validation sets
val_ratio = 0.10
split_idx = int(val_ratio * len(X2d))

np.random.seed(42)
indices = np.arange(len(X2d)) # len(X2d)=len(X6d)=len(X8d)
np.random.shuffle(indices)

x_train_X2d, x_val_X2d = X2d[indices[split_idx:]], X2d[indices[:split_idx]]
x_train_X6d, x_val_X6d = X6d[indices[split_idx:]], X6d[indices[:split_idx]]
x_train_X8d, x_val_X8d = X8d[indices[split_idx:]], X8d[indices[:split_idx]]
y_train, y_val = y[indices[split_idx:]], y[indices[:split_idx]]


# Print feature statistics
print("Computed " + str(X2d.shape[0]) + " features")
print("{} features for training, {} features for validation.".format(x_train_X2d.shape[0],x_val_X2d.shape[0]))
Y0 = [i for i, j in enumerate(y) if j == 0]
Y1 = [i for i, j in enumerate(y) if j == 1]
print("Class 0: " + str(len(Y0)) + " samples")
print("Class 1: " + str(len(Y1)) + " samples")
```

## Hyperparameter selection

```python
# find the best parameters for our logistic regression
penalties = [None, 'l1', 'l2', 'elasticnet']
class_weights = [None, 'balanced'] + [{0: 1, 1:i} for i in range(3, 7)]

best_params2d, f1_score2d = lin_reg_best_params(penalties, class_weights, x_train_X2d_norm, y_train)
print(f"Best parameters for 2d : {best_params2d}\n"
      f"f1 score for 2d : {f1_score2d}")

best_params6d, f1_score6d = lin_reg_best_params(penalties, class_weights, x_train_X6d_norm, y_train)
print(f"Best parameters for 2d : {best_params6d}\n"
      f"f1 score for 2d : {f1_score6d}")

best_params8d, f1_score8d = lin_reg_best_params(penalties, class_weights, x_train_X8d_norm, y_train)
print(f"Best parameters for 2d : {best_params8d}\n"
      f"f1 score for 2d : {f1_score8d}")
'''
# fit the model with the best parameters
logreg = linear_model.LogisticRegression(C=best_params["C"], penalty=best_params["penalty"], class_weight=best_params["class_weight"], solver=best_params["solver"], l1_ratio=best_params["l1_ratio"], max_iter=1000)
logreg.fit(x_train, y_train)
'''
```

```python
logreg2d = linear_model.LogisticRegression(C=best_params2d["C"], penalty=best_params2d["penalty"], class_weight=best_params2d["class_weight"], solver=best_params2d["solver"], l1_ratio=best_params2d["l1_ratio"], max_iter=1000)
logreg2d.fit(x_train_X2d_norm, y_train)

logreg6d = linear_model.LogisticRegression(C=best_params6d["C"], penalty=best_params6d["penalty"], class_weight=best_params6d["class_weight"], solver=best_params6d["solver"], l1_ratio=best_params6d["l1_ratio"], max_iter=1000)
logreg6d.fit(x_train_X6d_norm, y_train)

logreg8d = linear_model.LogisticRegression(C=best_params8d["C"], penalty=best_params8d["penalty"], class_weight=best_params8d["class_weight"], solver=best_params8d["solver"], l1_ratio=best_params8d["l1_ratio"], max_iter=1000)
logreg8d.fit(x_train_X8d_norm, y_train)

Z2d = logreg2d.predict(x_val_X2d_norm)
Z6d = logreg6d.predict(x_val_X6d_norm)
Z8d = logreg8d.predict(x_val_X8d_norm)

# Get non-zeros in prediction and grountruth arrays
Yn = np.nonzero(y_val)[0]

Z2dn = np.nonzero(Z2d)[0]
TPR2 = len(list(set(Yn) & set(Z2dn))) / float(len(Z2d))
print("True positive rate = " + str(TPR2))

Z6dn = np.nonzero(Z6d)[0]
TPR6 = len(list(set(Yn) & set(Z6dn))) / float(len(Z6d))
print("True positive rate = " + str(TPR6))

Z8dn = np.nonzero(Z8d)[0]
TPR8 = len(list(set(Yn) & set(Z8dn))) / float(len(Z8d))
print("True positive rate = " + str(TPR8))
```
